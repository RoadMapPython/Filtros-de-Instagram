```python
import numpy as np
import cv2 as cv
from scipy.interpolate import UnivariateSpline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

```python
pacote = 'sun.jpg'
```

```python
img=mpimg.imread(pacote)
```

```python
plt.rcParams['figure.figsize'] = [18,15]
plt.imshow(img)
```

```python
def mapping_function(x,y):
    spl=UnivariateSpline(x,y)
    return spl(range(256))
```

```python
def Quente(imagem):
    increase = mapping_function([0,64,128,192,256],[0,70,140,210,256])
    decrease = mapping_function([0,64,128,192,256],[0,40,90,150,256])
    red,green,blue =cv.split(imagem)
    red = cv.LUT(red,increase).astype(np.uint8)
    blue = cv.LUT(blue,decrease).astype(np.uint8)
    imagem = cv.merge((red,green,blue))
    return imagem

    
def  Frio(imagem) :
    increase = mapping_function([0,64,128,192,256],[0,70,140,210,256])
    decrease = mapping_function([0,64,128,192,256],[0,40,90,150,256])
    red,green,blue =cv.split(imagem)
    red = cv.LUT(red,decrease).astype(np.uint8)
    blue = cv.LUT(blue,increase).astype(np.uint8)
    imagem = cv.merge((red,green,blue))
    return imagem
```

```python
novafoto=Frio(img)
plt.imshow(novafoto)
```

```python
novafoto=Quente(img)
plt.imshow(novafoto)
```
